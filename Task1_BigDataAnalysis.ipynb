```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
df.info()
```

```python
df['Survived'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['Survived'].value_counts().plot(kind='bar')
plt.title("Titanic Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
df['Sex'].value_counts()
```

```python
df['Sex'].value_counts().plot(kind='bar')
plt.title("Male vs Female Passengers")
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='bar')
plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()
```

# Big Data Analysis – Insights


The dataset contains 891 Titanic passenger records.
Survival analysis shows that more passengers did not survive than survived.
Gender distribution analysis shows male passengers were more than female passengers.
Passenger class distribution shows most passengers traveled in Class 3.
Data analysis and visualization help understand patterns and trends in large datasets.
